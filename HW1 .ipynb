## OPENING AND READING THE TEXT FILE

```python
with open('text1.txt', 'r') as file:
    data = file.read()
```

```python
data
```

## Removing email addresses,Website links and URL's



```python
email = re.sub(r'\S+@\S+', '', data)
url = re.sub(r'http\S+', '', email)
weblinks = re.sub(r"www\S+", "",url)
ref = re.sub(r"\n","",weblinks)
num=re.sub(r'\d+', '', ref)
x = re.sub(r'\x0c', '', num)

```

```python
import string
final_data = re.sub(r'[^\w\s]', '', x)
```

```python
final_data
```

```python
final_data = final_data.lower()
```

```python
# Split the data using split() 
final_data1 = final_data.split()
final_data1
```

## Mapping

```python
mapped_words=[]
def map_func(final_data1):
    for i in range(0, len(final_data1)):
        if final_data1[i].lower() == final_data1[i-1].lower():
            mapped_words.append(final_data1[i] + '1')
        else:
            mapped_words.append(final_data1[i]+" " + str(1))
    
    return mapped_words
```

```python
map_func(final_data1)
```

## Shuffling

```python
sorted_words = sorted(mapped_words)
```

```python
sorted_words
```

## Reducing

```python
from collections import Counter
def reduce_func(sorted_words):
    word_counts = Counter(sorted_words)
    words = dict(word_counts)
    for word, count in word_counts.items():
        word = word.replace(" 1", "")
        print(f"{word}: {count}")
    return words
reduce_func(sorted_words)


```

```python
import nltk
import langid
from collections import Counter

with open('text2.txt', 'r', encoding='utf-8') as file:
    words = file.read()
    words = words.split()
    counter = Counter(words)
    
    nltk.download('words')
    english_words = set(nltk.corpus.words.words())
    
    non_english_words_counter = Counter()
    for word, count in counter.items():
        if word not in english_words and langid.classify(word)[0] != 'en':
            non_english_words_counter[word] += count
            
    non_english_words = dict(non_english_words_counter)
    print(non_english_words)
```
